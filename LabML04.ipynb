```python
#invocar biblioteca sklearn e importar modulo PCA e pandas
from sklearn.decomposition import PCA
import pandas as pd
```

```python
#ler um ficheiro csv e criar um df com os valores obtidos
df=pd.read_csv("https://raw.githubusercontent.com/masterfloss/data/main/worlddata.csv")
df
```

```python
#Mostrar os tipos de dados das variaveis
df.dtypes
```

```python
df['GDP']=pd.to_numeric(df['GDP'], downcast='float', errors='coerce')
```

```python
#Apagar a variavel Country
df1=df.drop(['Country'], axis=1)
```

```python
#remover todas as listas onde existem variaveis sem valores.
X=df1.dropna()
X
```

```python
#Criar um PCA que reduz o numero de features para 4
PCAModel = PCA(n_components=4)

#Transforma os features
XPCA = PCAModel.fit(X).transform(X)
XPCA
```

```python
#imprime o numero de fetures inicial e atual
print('Original number of features:', X.shape[1])
print('Reduced number of features:', XPCA.shape[1])
```

```python
#Mostra o racio explicado por cada variavel
PCAModel.explained_variance_ratio_
```
